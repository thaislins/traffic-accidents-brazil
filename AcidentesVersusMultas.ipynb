> Traffic accidents are a leading cause of death, injury and hospitalization, resulting in high economic and social costs. According to the [World Health Organization’s Global Report Status Report on Road Safety 2015](https://www.who.int/violence_injury_prevention/road_safety_status/2015/en/), more than 1.2 million people die every year on the roads around the world, most of them living in poor countries. And most of these traffic deaths are predictable and could be prevented with relative ease. Traffic infractions, which result in tickets/fines, are indications of bad roadside practices that could be avoided to prevent accidents, and it can be inferred that locations where the most traffic fines are issued are usually places where the most accidents happen.

> In Brazil, for example, despite the recent adoption of a series of measures to improve road safety, it was registered more than 46,000 deaths due to traffic accidents in 2013 . The Federal Highway Police (Polícia Rodoviária Federal - PRF) in Brazil attends about de 70,000km of federal highways and is distributed throughout the national territory, combating crime, assisting citizens, supervising, assessing and attending to accidents.

> In January 2017 the PRF started to use a new system to record the occurrence of traffic accidents and fines. We will use these and other datasets to investigate and obtain important information regarding traffic accidents in Brazil.


> The datasets can all be found in [Accidents - PRF](https://www.prf.gov.br/portal/dados-abertos/acidentes/acidentes) and [Fines - PRF](https://www.prf.gov.br/portal/dados-abertos/infracoes). The datasets of accidents by occurrence from 2007 to 2018 will be used to show a general description of the amount of accidents during these years. And the datasets pertaining accidents by ocurrence and person from 2017 will be analyzed in more detail. Other than that, the datasets that describe fines that happened in 2017 will be inspected and any possible correlations between this and the accidents dataset will be explored.


## 1.1 Describing all of the relevant datasets

The different datasets and their columns will be specified and described below


### 1.1.1 Accidents per occurence

* **id** Variable with numeric values, representing the identifier of the accident.

* **data_inversa** Date of occurrence in the format dd/mm/yyyy.

* **dia_semana** Day of the week of occurrence. Ex .: Monday, Tuesday, etc.

* **horario** Time of occurrence in the format hh:mm:ss.

* **uf** Federation unity. Ex .: MG, PE, DF, etc.

* **br** Variable with numeric values, representing the BR identifier of the accident.

* **km** Identification of the kilometer where the accident occurred, with a minimum value of 0.1 km and decimal case separated by a point.

* **municipio**  Name of the municipality where the accident occurred

* **causa_acidente** Identification of the main cause of the accident. In this data set, accidents with the main cause variable equal to "No" are excluded.

* **tipo_acidente** Identification of the type of accident. Example: Front collision, Track output, etc. In this set of data types of accidents with order greater than or equal to two are excluded. The order of the accident shows the chronological sequence of the types present in the same occurrence.

* **classificação_acidente** Classification regarding the severity of the accident: No Victims, With Victims Injured, With Victims Fatal and Ignored.

* **fase_dia** Stage of the day at the time of the accident. Ex. Dawn, Full day, etc.
sense

* **sentido_via** Direction of the road considering the point of collision: Crescent and descending.

* **condição_meteorologica** Weather conditions at the time of the accident: Clear sky, rain, wind, etc.

* **tipo_pista** Track type considering number of tracks: Double, single or multiple.

* **tracado_via** Description of track layout.

* **uso_solo** Description of the characteristics of the accident site: Urban = Yes; Rural = No.

* **latitude** Latitude of the accident site in decimal geodesic format.

* **longitude** Longitude of the accident site in decimal geodesic format.

* **pessoas**  Total number of people involved in the occurrence.

* **mortos**  Total dead people involved in the occurrence.

* **feridos_leves**  Total persons with minor injuries involved in the occurrence.

* **feridos_graves**  Total of people with serious injuries involved in the occurrence.

* **feridos**  Total number of injured persons involved in the occurrence (it is the sum of light and severe injuries).

* **ilesos**  Total number of uninvolved people involved in the occurrence.

* **ignorados**  Total number of people involved in the event and no known physical condition.

* **veiculos**  Total vehicles involved in the occurrence.

### 1.1.2 Accidents per person <a id='accid_person_description'></a>

This dataset has all the same columns as the ones in accidents by occurrences. Except it does **not** contain the columns **pessoas**, **feridos**, **ignorados**, **veiculos**. It contains a few different columns and ones with same name that have different meanings in this dataset, which are:

* **pesid** Variable with numeric values, representing the identifier of the person involved.

* **id_veiculo** Variable with numerical values, representing the identifier of the vehicle involved.

* **tipo_veiculo** Type of vehicle according to Art. 96 of the Brazilian Traffic Code. Ex .: Car, Truck, Motorcycle, etc.

* **marca** Description of the vehicle brand.

* **ano_fabricacao_veiculo** Year of manufacture of vehicle, yyyy format

* **tipo_envolvido** Type of involved in the accident according to participation in the event. Ex .: driver, passenger, pedestrian, etc.

* **estado_fisico** Condition of the involved according to the severity of the injuries. Example: dead, light wound, etc.

* **idade** Age of the involved. The code "-1" indicates that it was not possible to collect such information.

* **sexo** Sex of the involved. The "invalid" value indicates that it was not possible to collect such information.

* **ilesos** Binary value that identifies whether the involved was classified as uninjured.

* **feridos_leves** Binary value that identifies whether the involved was classified as lightly wounded.


* **feridos_graves** Binary value that identifies whether the involved was classified as seriously injured.


* **mortos** Binary value that identifies whether the deceased was classified as dead.

### 1.1.3 Traffic Infractions 

* **dat_infracao** Date of the infraction in the format dd / mm / yyyy

* **tip_abordagem** Identifies whether there was a vehicle approach: C (there was approach), S (no approach).

* **ind_assinou_auto** Variable that informs if the violator signed the notice of infraction. S (yes), Empty (no).

* **ind_veiculo_estrangeiro** Variable that informs if the vehicle is foreign. S (yes), N (no).

* **ind_sentido_trafego** Sense of traffic where the infraction occurred. C (increasing), D (decreasing).

* **uf_placa**  Vehicle plate federative unit.

* **uf_infracao** Federal unit of the place where the infraction occurred.

* **num_br_infracao** Variable with numerical values, representing the BR identifier where the infraction occurred.

* **num_km_infracao** Identification of the kilometer where the infraction occurred.

* **nom_municipio** Name of the municipality where the infraction occurred.

* **descricao_abreviada** Brief description of the infraction.

* **enquadramento** Framing of the infraction according to the CTB.

* **data_inicio_vigencia** Date in the dd/mm/yyyy format of the beginning of the infringement

* **data_fim_vigencia** Date in the dd/mm/yyyy format of the end of the infringement. When the infraction continues, the date will be the same as the last day of the month following the month with the last data available. Ex: the latest data available is from September/2016. The end date will be the same as 10/31/2016.

* **med_realizada** Record of the measurement carried out on radars, ethylometers, scales and trains.

* **med_considerada** Measurement considered for the record of the infraction.

* **exc_verificado** Excess verified in infractions where measuring equipment is used.

* **especie** Species of vehicle according to registration.

* **nome_veiculo_marca** Mark of the vehicle.

* **nom_modelo_veiculo** Vehicle model.

## 2.1 Reading data of accidents per occurence from 2007 to 2018 
> Let's start by reading the datasets corresponding to the accidents from 2007 to 2018. We will use the **Pandas** library to read these files.

> For this, we put the name of each file in a list, read one by one and save in an auxiliary dictionary each dataframe with the name of the file serving as a key.

> This way, it is easy to work with every year or only with a specific year desired.

```
import pandas as pd
filenames = [
    "datatran2007.csv",
    "datatran2008.csv",
    "datatran2009.csv",
    "datatran2010.csv",
    "datatran2011.csv",
    "datatran2012.csv",
    "datatran2013.csv",
    "datatran2014.csv",
    "datatran2015.csv",
    "datatran2016_atual.csv",
    "datatran2018.csv",
]

data_incidents = {}

for f in filenames:
  print(f)
  data_atual = pd.read_csv("{0}".format(f), encoding="Latin-1", sep=";", low_memory=False)
  key_name = f.replace(".csv", "")
  data_incidents[key_name] = data_atual
```

# 3.0 Data Cleaning

> These datasets we have chosen to analyze have a very wide variety of data, and a part of it doesn't have much useful information for us to investigate. For this reason we will analyze the datasets and their columns, as well as unique and unneeded values which won't lead us to any results. First of all, let's focus on what we want to analyze in each dataset so that we can remove any unnecessary columns. 

> Starting out with the columns in the datasets, as we can see in section 1, we have many columns, and we won't need all of this information, so let's find which ones we should be able to remove from each one.

## 3.1 Cleaning up the columns

> First of all, we will remove the columns which we are not going to use to analyze data in these datasets. And while doing that we will parse the dates contained in these datasets and join them with the time into one column, allowing us to remove the date (**data_inversa**) and time (**horario**) columns after that. 

```
# Cleaning columns in dataset of accidents per ocurrence 2017 and 2018
accid_occurrence = data_acid["acidentes_por_ocorrencia2017"]
accidents_2018 = data_incidents["datatran2018"]

# Parse dates in order to create a new column and remove two existing ones (data_inversa, horario)
accid_occurrence['data'] = pd.to_datetime(accid_occurrence['data_inversa'] + ' ' + accid_occurrence['horario'])
accidents_2018['data'] = pd.to_datetime(accid_occurrence['data_inversa'] + ' ' + accid_occurrence['horario'])

remove_columns = ["km", "sentido_via", "tipo_pista" , "tracado_via","ignorados", 
                  "uso_solo", "data_inversa", "horario", "delegacia", "regional", "uop"]

accid_occurrence = accid_occurrence.drop(columns=remove_columns)
accidents_2018 = accidents_2018.drop(columns=remove_columns)
accid_occurrence.info()
```

```
# Cleaning columns in dataset of accidents per person 2017 and 2018
accid_person = data_acid["acidentes_por_pessoa2017"]

# Parse dates in order to create a new column and remove two existing ones (data_inversa, horario)
accid_person['data'] = pd.to_datetime(accid_person['data_inversa'] + ' ' + accid_person['horario'])

remove_columns = ["km", "sentido_via", "tipo_pista" , "tracado_via","uso_solo", \
                  "marca", "ano_fabricacao_veiculo", "data_inversa", 
                  "horario", "delegacia", "regional", "uop"]

accid_person = accid_person.drop(columns=remove_columns)
```

## 3.1 Filtering column data

Now we will filter the values of latitude and longitude in the datasets, to make sure that there are no invalid locations, such as locations outside of Brazil. Other than that we will filter values in the columns of the dataset **accid_person** to remove invalid or insignificant values. As described in [section 1.1.2](#accid_person_description) there are some invalid values which we should extract to make the dataset smaller and cleaner.



```
# Filtering datasets of accidents per ocurrence 2017 and 2018 to keep location in Brasil

#lat < 5.65 || lat > -34 || long < -33.85 || long > -75.55

# Create a combined filter with most extreme points that delimit Brazil 
combined_filter_2018 = (accidents_2018['latitude'] < 5.65) & (accidents_2018['latitude'] > -34) & (accidents_2018['longitude'] < -33.85) & (accidents_2018['longitude'] > -75.55)
combined_filter_2017_occurrence = (accid_occurrence['latitude'] < 5.65) & (accid_occurrence['latitude'] > -34) & (accid_occurrence['longitude'] < -33.85) & (accid_occurrence['longitude'] > -75.55)
combined_filter_2017_person = (accid_person['latitude'] < 5.65) & (accid_person['latitude'] > -34) & (accid_person['longitude'] < -33.85) & (accid_person['longitude'] > -75.55)

# Use the filter on the dataframe
accidents_2018 = accidents_2018[combined_filter_2018]
accid_occurrence = accid_occurrence[combined_filter_2017_occurrence]
accid_person = accid_person[combined_filter_2017_person]
```

```
# Filtering values in dataset of accidents per person 2017 

# Remove age values which are invalid
remove_invalid_age = (accid_person['idade'] != -1)
accid_person = accid_person[remove_invalid_age]

# Remove sex values which are invalid
remove_invalid_values = (accid_person['sexo'] != 'Ignorado            ') & (accid_person['sexo'] != 'Não Informado       ')
accid_person = accid_person[remove_invalid_values]
```

## 4.3  Data from 2018 until October
The dataset that corresponds to 2018 is not complete: the data was accessed in December 2018, and the Federal Police had only made the data available until October of that year.

Even so, let's take a look at the panorama of this year in relation to traffic accidents.

### 4.3.1 Accidents by day
A very interesting fact that we can extract is what days have more accidents. We will verify this by separating accidents that have had victims (wounded or dead) and accidents that did not have victims.

```
import numpy as np

# Acidentes por dia da semana com vítimas
cond_victims = np.logical_or(np.logical_or(accidents_2018["mortos"] > 0, accidents_2018["feridos"] > 0), accidents_2018["feridos_graves"] > 0)
with_vitimas = accidents_2018[cond_victims]

values_cv = with_vitimas['dia_semana'].value_counts().keys().tolist()
counts_cv = with_vitimas['dia_semana'].value_counts().tolist()

# Acidentes por dia da semana sem vítimas
cond_without_victims = np.logical_and(np.logical_and(accidents_2018["mortos"] == 0, accidents_2018["feridos"] == 0), accidents_2018["feridos_graves"] == 0)
without_victims = accidents_2018[cond_without_victims]

values_sv = without_victims['dia_semana'].value_counts().keys().tolist()
counts_sv = without_victims['dia_semana'].value_counts().tolist()
```

```
p1 = plt.plot(counts_cv)

plt.title("Accidents on federal highways by day of the week in 2018, with victms")
plt.xlabel("Days")

plt.xticks(range(7), values_cv)

plt.show()
```

```
p1 = plt.plot(counts_sv)

plt.title("Accidents on federal highways by day of the week in 2018, no victms")
plt.xlabel("Days")

plt.xticks(range(7), values_sv)

plt.show()
```
